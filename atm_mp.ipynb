```python
def registration():
    
    while True:    
        fn=input("Enter your first name:")
        ln=input("Enter your last name:")
        city=input("Enter your city:")
        import random
        import getpass
        acc_num=random.randrange(1313100000,1313199999)

        username=input("Enter a username: ")
        a=collection.find()
        for i in a:
            if i.get("Username")==username:
                print("Username already exists! Please enter a defferent username")
                break
            else:
                continue

        pswrd1=getpass.getpass("Enter the password:")
        pswrd2=getpass.getpass("Enter the password again:")

        if pswrd1==pswrd2:
            print("Password set!")
            amount=float(input("Enter the amount: "))
            if amount>=4000:
                print("Done!")
                new_acc={"First_name":fn,"Last_name":ln,"City":city,"Account_Number":acc_num,"Username":username,"Password":pswrd1,"Amount":amount}
                collection.insert_one(new_acc)
                print("Your account is successfully opened, Welcome to AKA Bank!")
                break
            else:
                print("Insufficient balance, please add at least 4000")
        else:
            print("Passwords do not match")
            
def login():
    global rt1,rt2
    rt1 = ""
    rt2 = ""
    users = [list(collection.find())[i]["Username"] for i in range(len(list(collection.find())))]
    count=5
    while True:
        user = input("Enter the username: ")
        if user in users:
            while True:                        
                password=input("Enter your password: ")
                ind = users.index(user)

                if password==list(collection.find())[ind]["Password"]:
                    print("Login successful!")
                    rt1 = "done"
                    rt2 = user
                    break
                    
                elif count==1:
                    print("Wrong login credentials!!!")
                    break
                    
                else:
                    count=count-1
                    print("Wrong password. Remaining attempts:", count)
                    
            break

        elif count==1:
            print("Wrong login credentials!!!")
            break

        else:
            count=count-1
            print("Wrong username, Remaining attempts:", count)
        
    
def show_details(us):
    users = [list(collection.find())[i]["Username"] for i in range(len(list(collection.find())))]
    ind = users.index(us)
    amt = list(collection.find())[ind]["Amount"]
    fn = list(collection.find())[ind]["First_name"]
    ln = list(collection.find())[ind]["Last_name"]
    acc_num = list(collection.find())[ind]["Account_Number"]


    print("---------------- Account Details ---------------")
    print("Name: ",fn,ln)
    print("Account Number: ",acc_num)
    print("Balance: ",amt)    
    
def withdraw(us):
    users = [list(collection.find())[i]["Username"] for i in range(len(list(collection.find())))]
    ind = users.index(us)
    amt = list(collection.find())[ind]["Amount"]
    fn = list(collection.find())[ind]["First_name"]    

    while True:
        withdraw_amt = int(input("Enter the amount you want to withdraw: "))
        if withdraw_amt%100==0:
            if withdraw_amt<amt:
                amt = amt-withdraw_amt

                collection.update_one({"Username":us},{"$set":{"Amount":amt}})

                print("Successfully Withdrawn!!!")

                amt = list(collection.find())[ind]["Amount"]

                print("Account Balance: ",amt)
                break

            else:
                print("Insufficient balance")                

        else:
            print("Please put multiples of 100 Rs, 500Rs, 2000Rs")
            
def deposit(us):
    users = [list(collection.find())[i]["Username"] for i in range(len(list(collection.find())))]
    ind = users.index(us)
    amt = list(collection.find())[ind]["Amount"]
    fn = list(collection.find())[ind]["First_name"]    

    while True:
        deposit_amt = int(input("Enter the amount you want Deposit: "))
        if deposit_amt%100==0:
            amt = amt+deposit_amt

            collection.update_one({"Username":us},{"$set":{"Amount":amt}})

            print("Deposit successful!!!")

            amt = list(collection.find())[ind]["Amount"]

            print("Account Balance: ",amt)
            break

        else:
            print("Please put multiples of 100Rs, 500Rs, 2000Rs")
```
